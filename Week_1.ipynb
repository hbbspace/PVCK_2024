```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://cdn.popmama.com/content-images/community/20230519/community-dca5091bccaac63fab73ae950b6a6278.jpg"
        ]
#  baca dan tampilkan image
#  loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                                #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)       #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)        #convert color to RGB
  final_frame = cv.hconcat((image, image_2))            #concatenate image
  cv2_imshow(final_frame)                               #show image
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal di tengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Jawab: Resolusi gambar merujuk pada jumlah pixel yang ada dalam gambar, biasanya dalam bentuk lebar x tinggi (misalnya 1920x1080). Semakin tinggi resolusi, semakin banyak pixel yang membentuk gambar, sehingga gambar akan terlihat lebih detail. Gambar dengan resolusi rendah memiliki lebih sedikit pixel, sehingga setiap pixel terlihat lebih jelas dan gambar cenderung tampak kurang tajam atau pecah jika diperbesar. Pixel adalah elemen terkecil dari gambar yang mewakili satu titik warna, dan jumlah pixel menentukan kualitas visual keseluruhan dari gambar tersebut.


```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal di tengah image
for x in range (tinggi):
  image_3[x,int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis diagonal dari sudut kiri atas ke sudut kanan bawah
for x in range(tinggi):
    y = int((lebar / tinggi) * x)
    if y < lebar:
        image_3[x, y] = [255, 255, 255]

# Membuat garis diagonal dari sudut kiri bawah ke sudut kanan atas
for x in range(tinggi):
    y = int((lebar / tinggi) * (tinggi - x - 1))
    if 0 <= y < lebar:
        image_3[x, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# membuat garis horizontal secara acak
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(100, 200):
  image_3[int(tinggi/1.6), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

start = 90
for y in range (100):
    for c in range (20):
        image_3[start + c,start + y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
